```python
import numpy as np
import pandas as pd
import ast 
```

### TIỀN XỬ LÝ DỮ LIỆU

```python
post_df_full = pd.read_csv('Data/Asphalt9LegendsCleanData.csv')
```

```python
post_df_full['post_len'] = post_df_full['text'].apply(lambda x: len(x))

post_df_full
```

```python
# thêm cột tổng lượng tương tác của bài post (lượng reaction + lượng comment + lượng share)
post_df_full['total_interact'] = post_df_full['reaction_count'] + post_df_full['comments'] + post_df_full['shares']
```

```python
# thêm các cột thời gian đăng bài trong ngày và ngày đăng bài
post_df_full['time'] = pd.to_datetime(post_df_full['time'])
post_df_full['date'] = post_df_full['time'].dt.date
post_df_full['hour_in_day'] = post_df_full['time'].dt.time.apply(lambda x: x.hour)
```

```python
# Phan tich reactor
reactor_data = pd.DataFrame()
for i in range(1, len(post_df_full)):
    data_list = ast.literal_eval(post_df_full['reactors'].iloc[i]) 
    df = pd.DataFrame(data_list)
    reactor_data = pd.concat([reactor_data, df], ignore_index=True)

reactor_data.drop_duplicates(inplace= True)
reactor_data.to_csv("Data/Reactor_Data.csv", index=False)
```

```python
post_df_full['reactions'] = post_df_full['reactions'].apply(lambda x : dict(eval(x)))
post_df_full_reactions = post_df_full['reactions'].apply(pd.Series )

post_df_full_reactions
```

```python
# gộp các cột lại vào dataframe
post_df_full_with_reactions = pd.concat([post_df_full, post_df_full_reactions], axis=1).drop('reactions', axis=1)
post_df_full_with_reactions[['post_text','thích','yêu thích','wow','haha','thương thương','buồn','phẫn nộ',
                                 'shares','comments','reaction_count']]
```

```python
# Thay các dữ liệu NaN thành 0
post_df_full_with_reactions[['post_text','thích','yêu thích','wow','haha','thương thương','buồn','phẫn nộ',
                                 'shares','comments','reaction_count']].fillna(0)
```

```python
post_df_full_with_reactions.to_csv("Data/Post_df_full_with_reactions.csv", index=False)
```

```python
# Sữa lỗi thiếu dữ liệu ở cột comment_full
for i in range(0, len(post_df_full)):
    if (post_df_full['comments_full'].iloc[i][len(post_df_full['comments_full'].iloc[i])-1] != ']'):
        post_df_full['comments_full'].iloc[i] = post_df_full['comments_full'].iloc[i] + "]"
```

```python
def literal_eval_(comment):
    try:
        return ast.literal_eval(comment)
    except Exception as e:
        print("Error:", e)

import datetime

comment_data = pd.DataFrame()
        

for i in range(0, len(post_df_full)):
    comment_list = literal_eval_(post_df_full['comments_full'].iloc[i])
    comment_dt = pd.DataFrame(comment_list)
    comment_data = pd.concat([comment_data, comment_dt], ignore_index=True)

comment_data.to_csv("Data/Comment_Data.csv", index=False)
```
