### LÀM SẠCH DỮ LIỆU

```python
post_df_full = pd.read_csv('Data/Asphalt9Legends.csv')
```

```python
#Loại những dữ liệu không sử dụng hoặc bị lỗi trong quá trình cào
post_df_full.drop(['shared_text', 'post_text', 'original_text', 'image_lowquality', 'images_description', 'timestamp',
                   'images_lowquality_description', 'images_lowquality', 'video_duration_seconds', 'video_height', 'user_url',
                   'video_quality', 'video_size_MB', 'video_thumbnail', 'video_watches', 'video_width', 'factcheck', 'username',
                   'available', 'w3_fb_url', 'page_id', 'with', 'image_ids', 'video_ids', 'was_live', 'fetched_time', 'user_id',
                   'shared_post_url', 'shared_username', 'shared_user_id', 'shared_time', 'shared_post_id', 'sharers', 'is_live'], axis=1)
```

```python
# điền dãy ký tự trống cho các bài đăng không có tiêu đề
post_df_full['text'].fillna("", inplace=True)
post_df_full['post_text'].fillna("", inplace=True)

# xóa các hàng bị trùng lặp
post_df_full.drop_duplicates(inplace= True)

post_df_full['text']
```

```python
# xóa các hàng không có đủ dữ liệu 
indexAge = post_df_full[post_df_full['reactions'].isna()].index
post_df_full.drop(indexAge , inplace=True)
post_df_full.drop(238 , inplace=True)
post_df_full
```

```python
post_df_full.to_csv("Data/Asphalt9LegendsCleanData.csv", index=False)
```
